## 🐍 What is Python?
-> Python is a language for giving instructions to a computer.

Just like English lets us communicate with people, Python lets us communicate with computers. For example:

````python
print("hello world")
````

Means: "Computer, display these words on the screen."

## 🧰 Why use Python in biology?

Imagine you have measured the height of 10 plants. Easy. But what if you have measured 10,000 plants? Now suddenly:
- calculations become repetitive
- mistakes become likely
- plots become difficult

Computers excel at repetitive tasks, Python allows us to automate them. As you will see throughout this workshop, this language is very easy to learn, and allows us do to a whole range of cool things.

## 📦 What are packages?

Python itself is fairly small. When we import additional packages, we are adding tools. For example:

````python
import pandas
````

Means: "Give me access to the pandas toolbox."

````python
import matplotlib.pyplot as plt
````

Means: "Give me access to this plotting toolbox, and call it 'plt'," and

````python
import xarray as xr
````

Means: "Give me access to the toolbox for working with multidimensional scientific data, and call it 'xr'."

## 🔧 Why do we need to import things?

Is like asking: "Why doesn't every house come with every possible tool?"
Or like: "Why isn't every app from the App Store pre-installed on my phone?"

The answer is: most people don't need every tool, and it would be huge and inefficient. Python works similarly, we load only what we need. So packages are like toolboxes, and importing a package is like opening the toolbox you need.

## 🖥️ A few things about notebooks

At this very moment, you're working in a jupyter notebook. A notebook is just a sequence of cells you run one at a time, top to bottom. This one contains text, it's a `markdown` cell. You can see other cells that contain code, those are called `code` cells. You can add any type of cell you want above or below this cell by hovering your mouse over the top or bottom edge, and clicking the `+ Code` or `+ Markdown` symbol.

- **Running a cell**: click the ▶️ play button on the left of a cell, or click inside it and press **Shift+Enter**. Try re-running the `print("hello world")` cell above if you haven't yet.
- **Cells share memory**: anything you run in one cell (an import, a variable, a function) stays available to every cell below it — that's why we can `import pandas` in one cell and use `pandas` three cells later without importing it again. It also means *order matters*: if you run cells out of order, or skip one, things can break in confusing ways. If you ever get stuck, use the menu to "Run all" from the top and start clean.
- **Comments**: in Python code, anything after a `#` is a *comment* — the computer ignores it completely, it's just a note for humans. Scroll back up to the very first code cell (the setup cell) and look at all the `# ...` notes explaining each line. You can also put a `#` in front of a line of code to temporarily switch it off without deleting it (this is called *commenting out*) — you'll do this a lot over the next three days.
- **Ctrl+Z / Ctrl+F**: just like any text editor, **Ctrl+Z** undoes your last change inside a cell, and **Ctrl+F** searches the notebook for text — very handy once notebooks get longer.

## ⚙️ What is a function?

Basically, some code that does something for you. Let's look at an example:

````python
len("biology")
````

Can you guess what the function `len()` does? Try changing the text inside the quotes above and re-running the cell to test your guess.

Let's try another one: `mean()`. Unlike `len()`, it's not built into Python by default — it lives in a small toolbox called `statistics`. So, just like we did with `pandas`, `matplotlib`, and `xarray` a moment ago, we need to import it first:

````python
from statistics import mean
````

Now we can use it:

````python
mean([1, 2, 3])
````

Pretty obvious, no? `len()` and `mean()` are both **functions**: you give them some **input** in the parentheses (a piece of text, a list of numbers, ...) and they hand you back an **output**.

👉 Try `mean([10, 20, 30, 40])` yourself — either edit the cell above, or open a brand-new cell below it (hover just under a cell and click the small **+ Code** button that appears). Building the habit of making a throwaway "scratch" cell to test a quick idea, without touching the rest of the notebook, will save you a lot of confusion later on.

Some functions take more than one input. `round()` is a good example:

````python
round(3.14159, 2)
````

Two inputs separated by a comma (the number, and how many decimals to keep), one output. You'll see this `something(input1, input2, ...)` → output pattern everywhere in Python — including in the plot below:

````python
x = [1, 2, 3]
y = [2, 4, 6]
plt.plot(x, y)
````

Can you guess what we would call *x* and *y*?
What is *plt*?
What is *plot( )*?

### Answer:

*x* and *y* are what we call *variables*, *plt* is the *matplotlib.pyplot* toolbox we imported, and *plot* is the tool we're using to make a plot of *x* and *y*. `plot()` is a function too — same `input → output` pattern as `len()`, `mean()`, and `round()`, just with a picture as the output instead of a number. We'll do a lot more plotting like this on Day 1.

So, similar to how you would find a table of contents in a book:

```text
Biology textbook
└── Chapter 5
    └── Figure 3
```

We have the contents of toolboxes that we can import and use, e.g.:

```text
matplotlib
└── pyplot
    └── plot()
```

Very intuitive!

The last thing you need to know to get started with this workshop is that some things in Python contain both information (data) and useful actions. For example, when we open a dataset with the tool *xr.open_dataset( )*:

````python
my_dataset = xr.open_dataset('data/processed/land_sea_mask.nc')
````

*my_dataset* both contains information and useful functions we can use:

````python
print(list(my_dataset.data_vars))
my_dataset.land_sea_mask.plot()
````

## 👀 Always look at what you've got

Before doing anything else with a variable — plotting it, filtering it, computing something from it — get in the habit of just looking at it first: `print()` it, or if it's a table (a pandas DataFrame), use `.head()` to preview just the first few rows instead of dumping the whole thing on screen. It catches mistakes early, and it's the fastest way to build intuition for a new dataset. We already have `pandas` imported from earlier (remember — cells share memory!), so let's make one:

````python
fruit_prices = pandas.DataFrame({
    "fruit": ["mango", "banana", "papaya"],
    "price_php": [60, 15, 40],
})
fruit_prices.head()
````
