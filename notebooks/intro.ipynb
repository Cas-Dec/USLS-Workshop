## 🐍 What is Python?
-> Python is a language for giving instructions to a computer.

Just like English lets us communicate with people, Python lets us communicate with computers. For example:

```python
print("hello world")
```

Means: "Computer, display these words on the screen."

## 🧰 Why use Python in biology?

Imagine you have measured the height of 10 plants. Easy. But what if you have measured 10,000 plants? Now suddenly:
- calculations become repetitive
- mistakes become likely
- plots become difficult

Computers excel at repetitive tasks, Python allows us to automate them. As you will see throughout this workshop, this language is very easy to learn, and allows us do to a whole range of cool things.

## 📦 What are packages?

Python itself is fairly small. When we import additional packages, we are adding tools. For example:

```python
import pandas
```

Means: "Give me access to the pandas toolbox."

```python
import matplotlib.pyplot as plt
```

Means: "Give me access to this plotting toolbox, and call it 'plt'," and

```python
import xarray as xr
```

Means: "Give me access to the toolbox for working with multidimensional scientific data, and call it 'xr'."

## 🔧 Why do we need to import things?

Is like asking: "Why doesn't every house come with every possible tool?"
Or like: "Why isn't every app from the App Store pre-installed on my phone?"

The answer is: most people don't need every tool, and it would be huge and inefficient. Python works similarly, we load only what we need. So packages are like toolboxes, and importing a package is like opening the toolbox you need.
